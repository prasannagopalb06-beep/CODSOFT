```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

movies={ 'movie':['Inception','Titanic','Avengers','Interstellar','Avatar','Iron Man','Joker','The Matrix','Gladiator','Notebook','Vikram','Leo','Master','Kaithi','Asuran','Jai Bhim','Soorarai Pottru','96','Visaranai','Enthiran' ],
        'genre':['Action Sci-Fi','Romance Drama','Action Fantasy','Sci-Fi Drama','Sci-Fi Fantacy','Action Sci-Fi','Psychological Drama','Action Sci-Fi','Action Drama','Romance','Action Thriller','Action Thriller','Action Drama','Action Thriller','Drama Thriller','Legal Drama','Inspirational Drama','Romance Drama','Crime Drama','Sci-Fi Action']}
movie_df=pd.DataFrame(movies)

ratings={
    'User': ['A','A','A','B','B','B','C','C','D','D','E','E'],
    'Movie': ['Inception','Titanic','Vikram','Avengers','Iron Man','Master','96','Notebook','Joker','Asuran','Kaithi','Soorarai Pottru'],
    'Rating': [5,2,5,5,5,4,5,4,5,5,5,5]
    }
rating_df=pd.DataFrame(ratings)

vectorizer=CountVectorizer()
genre_matric=vectorizer.fit_transform(movie_df['genre'])
similarity=cosine_similarity(genre_matric)

def content_based(movie_name):
  index=movie_df[movie_df['movie']==movie_name].index[0]
  scores=list(enumerate(similarity[index]))
  scores=sorted(scores,key=lambda x:x[1],reverse=True)[1:3]
  return [movie_df.iloc[i[0]].movie for i in scores]

user_item=rating_df.pivot_table(index='User',columns='Movie',values='Rating')
def collaborative(user):
  user_movies=user_item.loc[user].dropna().index
  similar_user=user_item.drop(user)
  recommendations=[]
  for u in similar_user.index:
    movies=user_item.loc[u].dropna().index
    for m in movies:
      if m not in user_movies:
        recommendations.append(m)
  return list(set(recommendations))

def hybrid_recommend(user,movie):
  cb=content_based(movie)
  cf=collaborative(user)
  final_recommandation=list(set(cb+cf))
  return final_recommandation

print("Hybrid Recommendations:")
print(hybrid_recommend('A','Inception'))
```
