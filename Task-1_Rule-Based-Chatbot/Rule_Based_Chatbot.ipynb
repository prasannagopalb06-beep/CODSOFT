```python
import re
import gradio as gr
from datetime import datetime

rules={
    r"\b(hi|hello|hey)\b":"Hello! How can i help you?",
    r"\b(how are you)\b":"I am doing great! Thanks for asking.",
    r"\b(your name|who are you)\b":"I am a rule-based chatbot built using a python.",
    r"\b(help|support)\b":"you can ask about time,date,college,food,hostel or general question.",
    r"\b(time)\b":f"current time is {datetime.now().strftime('%H:%M:%S')}",
    r"\b(date|day)\b":f"Today's date is {datetime.now().strftime('%d-%m-%y')}",

    r"\b(college)\b":"This chatbot can assist with basic college-related queries.",
    r"\b(department)\b": "We offer Computer Science, AI, ECE, and Mechanical departments.",
    r"\b(hostel)\b": "Hostel facilities are available for both boys and girls.",
    r"\b(canteen|food)\b": "Canteen provides veg and non-veg food at affordable prices.",
    r"\b(library)\b": "Library is open from 9 AM to 8 PM.",
    r"\b(exam)\b": "Please check the academic calendar for exam schedules.",
    r"\b(fees)\b": "Fee details are available in the college office.",
    r"\b(placement)\b": "Our placement cell supports students with training and interviews.",


    r"\b(ai|artificial intelligence)\b": "AI enables machines to think and learn like humans.",
    r"\b(machine learning)\b": "Machine Learning is a subset of AI that learns from data.",
    r"\b(python)\b": "Python is a popular programming language used in AI and web development.",
    r"\b(chatbot)\b": "A chatbot is a program that interacts with users using text or voice.",
    r"\b(weather)\b": "I can't fetch live weather, but you can check online ",


    r"\b(thank you|thanks)\b": "You're welcome! ",
    r"\b(sorry)\b": "No problem at all ",


    r"\b(bye|exit|quit)\b": "Goodbye!  Have a great day!"
}

fallback="Sorry,I didn't understand that Please try something else."

def chatbot(user_input):
  user_input=user_input.lower()

  for pattern,response in rules.items():
    if re.search(pattern,user_input):
      return response
  return fallback

interface=gr.Interface(fn=chatbot,inputs=gr.Textbox(lines=2,placeholder="Type your message here...."),outputs="text",title="Rule-Based Chatbot",description="A simple rule-based chatbot using python,Regex and Gradio")
interface.launch()
```
